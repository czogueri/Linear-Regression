```python
#Train and Test data
```

```python
data.columns
```

```python
y = data['Salary']
x = data['YearsExperience']
```

```python
#use sklearn to import Ml dependency
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=584)
```

```python
#Train Model by importing linearRegression model from sklearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#create instance of linear regression model
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
#to predict the test results
```

```python
y_red = lm.predict(X_test)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
#Time to predict data
```

```python
plt.scatter(y_test,y_red)
plt.xlabel('YearsExperience')
plt.ylabel('Salary ($)')
```

```python
#evaluate models peformance using metrics
```

```python
from sklearn import metrics
```

```python
#use metrics dependency to find mean absolute error (MAE), mean square error (MSE), and Root mean squared error (RMSE)
```

```python
print('MAE', metrics.mean_absolute_error(y_test, y_red))
print('MSE', metrics.mean_squared_error(y_test, y_red))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test, y_red)))
```

```python
#to test accuracy use score function
```

```python
print(lm.score(X_train, y_train))
```

```python
metrics.explained_variance_score(y_test,y_red)
```
