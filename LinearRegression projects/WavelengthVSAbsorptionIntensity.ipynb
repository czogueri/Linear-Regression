```python
#train and test data
```

```python
X =data['Wavelength ']
y = data['Absorption Intensity']
```

```python
#use sklearn to import machine learning dependencies
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#time to train model by importing the LRegression model from sklearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#create instance of lineafr regression model
```

```python
lm = LinearRegression()
```

```python
lm.fit(X=(X.array.reshape(-1, 1)),y= y)
```

```python
#to predict the test results
X = X_test
```

```python
lm.coef_
```

```python
print(lm.intercept_)
```

```python
#Time to predict data
```

```python
y_red = lm.predict(X_test.array.reshape(-1,1))
```

```python
print( X_train.shape)
print( y_train.shape)
print( y_red.shape)
print(y_test.shape)
```

```python
plt.scatter(y_test, y_red)
plt.title('Wavelength VS Absorption Intensity')
plt.ylabel('Absorption Intensity (A.U)')
plt.xlabel('Wavelength (nm)')
```

```python
#evaluate model peformance using metrics
```

```python
from sklearn import metrics 
```

```python
#use metrics dependencies to determine the Mean Absolute Error (MAE), Mean Square Error (MSE), and Root Mean Square Error (RMSE)
```

```python
print('MAE', metrics.mean_absolute_error(y_test, y_red))
print('MSE', metrics.mean_squared_error(y_test, y_red))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, y_red)))
```

```python
#to test accuracy use score module
```

```python
print (lm.score(X_train.array.reshape(-1,1), y_train))
#print( X_test.shape)
```
